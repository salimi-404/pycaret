# PyCaret Clustering on Kiva Dataset

```python
from pycaret.datasets import get_data
from pycaret.clustering import *

df = get_data('kiva')
df.head()
```

```python
s = setup(df, normalize=True, session_id=123)
```

```python
best = create_model('kmeans')
best
```

```python
plot_model(best, plot='cluster')
```
